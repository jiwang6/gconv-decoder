```python

import matplotlib.pyplot as plt

import numpy as np

Ls = [5,7,9]
ps = [155, 166, 178, 180]


for L in Ls:
    for p in ps:
        with open(f'f1-scores/f1_L_{L}_p_{p}.txt') as file_object:
            # read until the end of the file
            contents = file_object.read()

        score_arrys = contents.split(']')

        score_arrys = score_arrys[:-1]

        new_arry = []

        for arry in score_arrys:
            # strip [
            arry = arry[1:]

            # convert to list
            arry = arry.split(',')

            # convert to numpy float
            arry = [float(item) for item in arry]

            arry = np.array(arry)

            new_arry.append(arry)

        new_arry = np.array(new_arry)

        # box plot transposed
        plt.boxplot(new_arry)
        plt.title(f'F1 scores for L={L}, p={p}')
        plt.xlabel('Decoding')
        plt.ylabel('F1 score')

        # make x ticks from 0 to 15
        plt.xticks(np.arange(1, 17), np.arange(0, 16))


        # save plot
        plt.savefig(f'f1-scores/f1_L_{L}_p_{p}.png')

        plt.show()


```
